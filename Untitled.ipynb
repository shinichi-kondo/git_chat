```python
import numpy as np
from matplotlib import pyplot as plt
 
def is_inner(x, y):
    """
    円の内側かどうかを判定
    """
    return x ** 2 + y ** 2 < 1
 
 
inner_points_cnt = 0
all_points_cnt = 0
 
 
X, Y = np.random.rand(2, 10 ** 5)
for x, y in zip(X, Y):
    all_points_cnt += 1
    if is_inner(x, y):
        inner_points_cnt += 1
 
pi = (inner_points_cnt / all_points_cnt) * 4
 
# 計算結果を表示
print(pi)
 
 
# 以下matplotlibによる可視化
 
# ランダムに配置した点
plt.scatter(X, Y, s=1)
 
# x ** 2 + y ** 2 = 1 のグラフ
cx = np.linspace(0, 1, 100)
cy = np.sqrt(np.subtract(1, cx ** 2))
plt.plot(cx, cy, linewidth=3, color='r')
 
# 縦横比をそろえる
plt.gca().set_aspect('equal', adjustable='box')
 
# グラフの表示
plt.show()
```
